# 크롬드라이버로 사이트 접속

```python
# 크롬드라이버로 사이트 연결
from selenium import webdriver as wb
driver = wb.Chrome(executable_path="chromedriver.exe")
```

```python
driver.get("https://www.acmicpc.net/")
```

```python
body_tag = driver.find_element_by_tag_name("body")
body_tag
```

```python
from selenium.webdriver.common.keys import Keys
```

```python
#메뉴버튼 클릭
driver.find_element_by_class_name("navbar-toggle").click()
```

```python
#문제 드롭다운버튼 클릭
driver.find_element_by_class_name("dropdown-toggle").click()
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
```

```python
#단계별로 문제풀기 클릭
WebDriverWait(driver,10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "a[href='/step']")))
```

```python
WebDriverWait(driver,10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "a[href='/step']"))).click()
```

```python
#page소스를 page에 대입
page = driver.page_source
page
```

```python
from bs4 import BeautifulSoup
#page에 저장된 태그를 beautifulSoup을 이용해 정리해 리턴
soup = BeautifulSoup(page,'lxml')
soup
```

```python
# 제목만 뽑아오기
problems_subject = [href.text for href in soup.find("div", {"class" : "table-responsive"}).find_all("a")] 
problems_subject
```

```python
problems= [href.text for href in soup.find("div", {"class" : "table-responsive"}).find_all("td")] 
problems
```

```python
level=[]
subject=[]
explain=[]
number=[]
for i in range(0,len(problems)):
    if i%6 ==0:
        level.append(problems[i])
    elif i%6 ==1:
        subject.append(problems[i]) 
    elif i%6 ==2:
        explain.append(problems[i]) 
    elif i%6 ==4:
        number.append(problems[i])
print(level)
print(subject)
print(explain)
print(number)
```

```python
import pandas as pd
from pandas import DataFrame as df
```

```python
table = df(data={'level':level, '제목':subject,'설명':explain,'문제개수':number})
```
