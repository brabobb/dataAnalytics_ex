```python
import csv
import matplotlib.pyplot as plt
import random
import math
```

```python
#제주도 남녀 인구 꺾은선 그래프로 조회

#파일 오픈
f = open('gender.csv')
data = csv.reader(f)

female = []
male = []

#원하는 지역 입력 받기
addr = input('찾고 싶은 지역의 이름을 알려주세요: ')

for row in data:
    if addr in row[0]:
        for i in range(101):
            male.append(int(row[i+3]))
            female.append(int(row[-(i+1)]))
        break
female.reverse()

plt.plot(male, label = 'Male')
plt.plot(female, label = 'Female')
plt.rcParams["figure.dpi"] = 300
plt.legend()
plt.show()
```

```python
#제주도 남녀 인구 차이 알아보기
result = []

#원하는 지역 입력 받기
addr = input('찾고 싶은 지역의 이름을 알려주세요: ')

for row in data:
    if addr in row[0]:
        for i in range(3, 104):
            #남성인구 - 여성인구
            result.append(int(row[i]) - int(row[i+103]))
        break

#bar(x축, y축)
plt.bar(range(101), result)
plt.show()
```

```python
#random 함수 할용한 위치, 크기 서로 다른 100개의 점 
x = []
y = []
size = []

for i in range(100):
    x.append(random.randint(50, 100))
    y.append(random.randint(0, 50))
    size.append(random.randint(0, 100))

#alpha = 투명도 (0일수록 투명)
plt.scatter(x, y, s = size, c = range(100), alpha = 0.5)
plt.colorbar()
plt.show()
```

```python
#파일 오픈
f = open('gender.csv')
data = csv.reader(f)

female = []
male = []
size = []

#원하는 지역 입력 받기
addr = input('찾고 싶은 지역의 이름을 알려주세요: ')

for row in data:
    if addr in row[0]:
        for i in range(3, 104):
            male.append(int(row[i]))
            female.append(int(row[i+103]))
            size.append(math.sqrt(int(row[i]) + int(row[i+103])))
        break

plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.figure(figsize = (10, 5), dpi = 300)

plt.title(addr + '지역의 성별 인구 그래프')
plt.scatter(male, female, s = size, c = range(101), alpha = 0.5, cmap = 'jet')
plt.colorbar()
plt.plot(range(max(male)), range(max(male)), 'g')

plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
```
