# UNIT 10 우리동네 인구 구조를 산점도로 나타내기

```python
# 입력받은 구의 남녀 인구 꺾은선그래프 그리기
import csv
import matplotlib.pyplot as plt
f=open('C:/Users/Admin/Downloads/gender.csv','r', encoding='cp949')
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
data = csv.reader(f)
male = []
female = []
name = input('궁금한 동네를 입력해주세요 : ')
for row in data :
    if name in row[0] :
        for i in range(3,104) :
            male.append(int(row[i+103].replace(',','')))    # 천의자리 구분용 쉼표 삭제
            female.append(int(row[i+206].replace(',','')))
        break
plt.plot(male, label='남성')
plt.plot(female, label='여성')
plt.legend()
plt.show()
```

```python
# 연령대별 인구차(남성 인구- 여성 인구)를 막대그래프로 표현하기
result = [] 
name = input('궁금한 동네를 입력해주세요 : ')
for row in data :
    if name in row[0] :
        for i in range(3,104) :
            result.append(int(row[i+103])-int(row[i+206]))
        break
plt.bar(range(101),result)
plt.show()
```

```python
# 간단한 산점도 그리기
plt.style.use('ggplot') # 격자무늬 스타일 지정하기
plt.scatter([1,2,3,4], [10,30,20,40])
plt.show()
```

```python
# 버블 차트 그리기
# 왼쪽 리스트 : x값, 오른쪽 리스트 : y값, s : 점들의 크기, c : 표현하고자 하는 색상 수
plt.scatter([1,2,3,4], [10,30,20,40], s=[100,200,250,300], c=range(4))
plt.colorbar()
plt.show()
# 기타 다양한 컬러맵은 https://matplotlib.org/tutorials/colors/colormaps.html
```

```python
# 랜덤함수를 이용한 버블 차트 그리기
import matplotlib.pyplot as plt
import random
x = []
y = []
size = []
for i in range(100) :
    x.append(random.randint(50,100))
    y.append(random.randint(50,100))
    size.append(random.randint(10,100))
# cmap : 컬러맵 적용, alpha : 투명도 조정(0으로 갈수록 투명, 1로 갈수록 불투명)
plt.scatter(x, y, s=size, c=size, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```

```python
# 입력받은 구의 남녀 인구 버블 차트 그리기
import math
f=open('C:/Users/Admin/Downloads/gender.csv','r', encoding='cp949')
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
data = csv.reader(f)
male = []
female = []
size = [] # 점(버블)의 크기를 결정할 리스트
name = input('궁금한 동네를 입력해주세요 : ')
for row in data :
    if name in row[0] :
        for i in range(3,104) :
            male.append(int(row[i+103].replace(',','')))
            female.append(int(row[i+206].replace(',','')))
            size.append(math.sqrt(int(row[i+103].replace(',',''))+int(row[i+206].replace(',',''))))
        break
# sqrt() : 제곱근 값 계산
# 이 메소드를 쓴 이유는 안쓰면 점의 크기가 너무 커져 가독성을 해치기 때문
plt.figure(figsize=(10,5), dpi=300)
plt.title(name+' 지역의 성별 인구 그래프')
plt.scatter(male, female, s=size, c= range(101), alpha=0.5, cmap='jet')
plt.colorbar()
plt.plot(range(max(male)),range(max(male)), 'g')
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
f.close()
```
