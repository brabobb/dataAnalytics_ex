```python
#날짜 형태로 된 8개의 인덱스
index = pd.date_range('1/1/2000', periods=8)
print(index)
```

```python
# 8행 3열로 구성된 랜덤 데이터를 생성
# 인덱스와 컬럼 이름을 정하여 데이터 프레임으로
import numpy as np
df = pd.DataFrame(np.random.rand(8,3), index=index, columns=list('ABC'))
df
```

```python
#데이터 프레임에서 특정 행이나 열을 선택하면 
#인덱스와 1차원 배열 형태의 데이터로 이루어진 시리즈(Series)라는 데이터 구조 형태로 표현
index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8,3), index=index, columns=list('ABC'))
print(df['B'])
```

```python
# 마스크 기능을 데이터 프레임에서도 사용
index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8,3), index=index, columns=list('ABC'))
print(df['B'] > 0.4)
```

```python
#B열의 값이 0.4보다 큰 값이라는 조건이 True인 데이터 프레임
df2 = df[df['B'] > 0.4]
df2
```

```python
# 열 방향을 축으로 계산하려면 axis를 1로 설정
# axis = 0은 행 방향을 나타내며 0이 기본 값
```

```python
#전체 행에 대해서 A열의 값을 B열의 값으로 나눈 후, 그 결과를 새로 만든 D열에 저장
import pandas as pd
import numpy as np
index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8,3), index=index, columns=list('ABC'))
df['D'] = df['A'] / df['B']  # A열의 값을 B열의 값으로 나눈 값을 D열에 저장
df
```

```python
# head() 함수는 많은 데이터 중 처음 5개의 데이터만 확인하고 싶을 때 사용
import pandas as pd
import numpy as np
 
index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8,3), index=index, columns=list('ABC'))
df['D'] = df['A'] / df['B']
df['E'] = np.sum(df, axis=1)  # 행 우선 계산 값을 E열에 저장
df.head()
```

```python
#열 우선계산
import pandas as pd
import numpy as np
 
index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8,3), index=index, columns=list('ABC'))
df['D'] = df['A'] / df['B']
df['E'] = np.sum(df, axis=1)
df = df.sub(df['A'], axis=0)  # A열의 데이터를 기준으로 열 우선 계산
df.head()
```

```python
# 데이터 전체를 C열 값으로 나눌 때는 div() 함수
import pandas as pd
import numpy as np
 
index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8,3), index=index, columns=list('ABC'))
df['D'] = df['A'] / df['B']
df['E'] = np.sum(df, axis=1)
df = df.sub(df['A'], axis=0)
df = df.div(df['C'], axis=0)    # C열 데이터를 기준으로 열 우선 계산
df.to_csv('test.csv')           # 데이터 프레임을 test.csv 파일로 저장
df.head()
```
