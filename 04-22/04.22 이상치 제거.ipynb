```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
train=pd.read_csv("train.csv",sep=",",dtype='unicode')
plt.rc('font',family='Malgun Gothic') # 한글 폰트 지원
plt.rcParams['axes.unicode_minus'] = False # 그래프의 음수 부호가 깨지지 않음

train
```

```python
# 폐업한 병원의 16년도 당기순이익을 막대그래프로 출력
f=open('train.csv','r', encoding='cp949')
data=csv.reader(f)
sixteen=[] # 16년도 당기순이익
seventeen=[] # 17년도 당기순이익
hos_name=[] # 폐업한 병원의 ID번호
for row in data:
    if row[1] == ' close' : 
        hos_name.append(row[0])
        sixteen.append(int(row[39]))
        seventeen.append(int(row[15]))
print(hos_name)
plt.title('폐업한 병원의 2016, 2017년 당기순이익')
plt.bar(np.arange(len(hos_name)), sixteen, alpha=0.5, color='b')
plt.bar(np.arange(len(hos_name)), seventeen, alpha=0.5, color='r')
plt.xticks(np.arange(len(hos_name)), hos_name)
plt.ylim(-3000000000,3000000000)
plt.show()
f.close()
# 209, 265, 306, 366, 414, 423, 427번 병원의 당기순이익 수치의 폭이 매우 크므로 이들을 따로 분류
```

```python
# 이상치 제거 함수
def outliers_iqr(data):
    q1,q3=np.percentile(data,[25,75])
    iqr=q3-q1
    lower_bound=q1-(iqr*1.5)
    upper_bound=q3+(iqr*1.5)
    return np.where((data>upper_bound)|(data<lower_bound))

sixteen_outlier=outliers_iqr(sixteen)[0]   # 2016년 당기순이익 데이터 중 이상치를 갖는 hos_name
seventeen_outlier=outliers_iqr(seventeen)[0]    # 2017년 당기순이익 데이터 중 이상치를 갖는 hos_name
print('2016년 당기순이익 데이터 중 이상치를 갖는 폐업 병원ID :',sixteen_outlier)
print('2017년 당기순이익 데이터 중 이상치를 갖는 폐업 병원ID :',seventeen_outlier)
```

```python
not_good_hos=[] # 이상치의 당기순이익 데이터를 갖는 폐업 병원의 ID번호
new_hos=[] # 정상적인 당기순이익 데이터를 갖는 폐업 병원의 ID번호

print('폐업한 병원의 ID번호 :',hos_name)

# not_good_hos에 2016년도 이상치를 갖는 병원ID 추가 
for i in range(0,len(sixteen_outlier)):  
    not_good_hos.append(hos_name[sixteen_outlier[i]]) # 폐업 병원의 ID 중 2016년 데이터가 이상한 병원의 ID 정보를 리스트에 추가

# not_good_hos에 2017년도 이상치를 갖는 병원ID 추가 
for i in range(0,len(seventeen_outlier)):
    if hos_name[seventeen_outlier[i]] not in not_good_hos: # 2016년도 이상치를 갖는 병원ID 중복 방지
        not_good_hos.append(hos_name[seventeen_outlier[i]]) # 중복 제외 2017년 데이터가 이상한 병원의 ID 정보를 리스트에 추가
not_good_hos.sort()  # 이상치를 내놓는 병원ID 값을 오름차순으로 정렬
print('이상치의 당기순이익 데이터를 갖는 폐업 병원의 ID번호 :',not_good_hos)

# 폐업 병원 중 이상치가 없는 병원ID 추가 
for i in range(0,len(hos_name)):
    if hos_name[i] not in not_good_hos: # not_good_hos의 정보를 제외한 병원ID를 new_hos에 추가
        new_hos.append(hos_name[i])
print('정상적인 당기순이익 데이터를 갖는 폐업 병원의 ID번호 :',new_hos)
```

```python
# 폐업한 병원의 16,17년도 당기순이익을 막대그래프로 출력 (이상치 제거)
f=open('train.csv','r', encoding='cp949')
data=csv.reader(f)
fixed2016=[] # 이상치를 제거한 2016년 당기순이익 자료
fixed2017=[] # 이상치를 제거한 2017년 당기순이익 자료
for row in data:
    if row[1] == ' close' and row[0] in new_hos : 
        fixed2016.append(int(row[39]))
        fixed2017.append(int(row[15]))
        
plt.title('폐업한 병원의 2016, 2017년 당기순이익(이상치 제거)')
plt.bar(np.arange(len(new_hos)), fixed2016, alpha=0.5, color='b')
plt.bar(np.arange(len(new_hos)), fixed2017, alpha=0.5, color='r')
plt.xticks(np.arange(len(new_hos)), new_hos)
plt.ylim(-300000000,300000000)
plt.show()
f.close()
```
