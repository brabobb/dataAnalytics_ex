# UNIT 07 우리동네 인구 구조 시각화하기

```python
# 우리 구의 동별 총인구 그래프 그리기
import csv
import matplotlib.pyplot as plt
f=open('C:/Users/user/Downloads/age.csv','r', encoding='cp949') # 파일 디렉토리를 알고있다면 바로 불러오기 가능
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False

data=csv.reader(f)
result=[]
for row in data:
    if '인천광역시 중구' in row[0] : 
        result.append(int(row[1].replace(',',''))) # 값이 쉼표가 들어간 문자열이므로 쉼표를 없애고 정수로 변환
result.remove(result[0]) # 총인구에 해당하는 첫 데이터는 삭제

plt.title('인천광역시 중구의 동별 총인구 그래프')
plt.style.use('ggplot')
plt.plot(result)
plt.show()
f.close()
```

# UNIT 08 인구 구조를 다양한 형태로 시각화하기

```python
# 입력받은 동의 총인구 그래프 그리기
import csv
import matplotlib.pyplot as plt
f=open('C:/Users/user/Downloads/age.csv','r', encoding='cp949')
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False

data=csv.reader(f)
result=[]
name=input('인구 구조가 알고 싶은 지역의 이름(구 단위)을 입력해주세요 : ')
for row in data:
    if name in row[0] : 
        result.append(int(row[1].replace(',','')))
result.remove(result[0])

plt.title(name+'의 동별 총인구 그래프')
plt.style.use('ggplot')
plt.plot(result)
plt.show()
f.close()
```

```python
# 우리 구의 남녀인구 항아리 그래프 그리기
f=open('C:/Users/user/Downloads/gender.csv','r', encoding='cp949')
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
data=csv.reader(f)
male_result=[]
female_result=[]
for row in data:
    if '인천광역시 중구' in row[0] : 
        male_result.append(-int(row[4].replace(',','')))
        female_result.append(int(row[5].replace(',','')))

male_result.remove(male_result[0])
female_result.remove(female_result[0])
# print(male_result)
# print(female_result)

plt.title('인천광역시 중구의 동별 남녀인구 그래프')
plt.barh(range(len(male_result)),male_result,label='남성')
plt.barh(range(len(female_result)),female_result,label='여성')
plt.legend()
plt.show()
f.close()
```

# UNIT 09 우리 동네 인구구조를 파이 차트로 나타내기

```python
f=open('C:/Users/user/Downloads/gender.csv','r', encoding='cp949')
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
data=csv.reader(f)
male_result=[]
female_result=[]
name=input('인구 구조가 알고 싶은 지역의 이름(구 단위)을 입력해주세요 : ')
for row in data:
    if name in row[0] : 
        male_result.append(-int(row[4].replace(',','')))
        female_result.append(int(row[5].replace(',','')))

male_result.remove(male_result[0])
female_result.remove(female_result[0])
# print(male_result)
# print(female_result)

plt.title(name+'의 남녀인구 그래프')
plt.barh(range(len(male_result)),male_result,label='남성')
plt.barh(range(len(female_result)),female_result,label='여성')
plt.legend()
plt.show()
f.close()
```

```python
# 간단한 파이 차트 그리기
plt.pie([10, 20])
plt.show()
```

```python
# 파이 차트의 레이블, 색, 반지름, 비율 자동 계산 이용하기
plt.rc('font', family='Malgun Gothic')
size = [2441, 2312, 1031, 1233]
label = ['A형','B형','AB형', 'O형']
color = ['darkmagenta', 'deeppink', 'hotpink', 'pink']
# matplotlib에서 쓸 수 있는 색 목록 : https://matplotlib.org/gallery/color/named_colors.html
plt.axis('equal')
plt.pie(size, labels=label, autopct='%.1f%%', colors=color, explode=(0,0,0.1,0))
# autopct : 비율을 자동으로 계산하는 파이 차트의 속성
# explode : 파이의 특정 조각을 정해진 수치만큼 돌출시키는 파이 차트의 속성
plt.legend()
plt.show()
```

```python
# '제주특별자치도의 남녀 총인구 파이 차트
f=open('C:/Users/user/Downloads/gender.csv','r', encoding='cp949')
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
data=csv.reader(f)
male_result=[]
female_result=[]
for row in data:
    if '제주특별자치도 제주시' in row[0] : 
        male_result.append(int(row[4].replace(',','')))
        female_result.append(int(row[5].replace(',','')))

ml=0
fl=0
male_result.remove(male_result[0])
female_result.remove(female_result[0])

for i in range(len(male_result)): # range 안에 female_result를 써도 무관
    ml+=male_result[i]
    fl+=female_result[i]
size=[]
size.append(ml)
size.append(fl)

color = ['crimson', 'darkcyan']
plt.axis('equal')
plt.pie(size, labels=['남', '여'], autopct='%.1f%%', colors=color, startangle=90)
plt.title('제주특별자치도의 남녀 총인구 파이 차트')
plt.legend()
plt.show()
f.close()
```
