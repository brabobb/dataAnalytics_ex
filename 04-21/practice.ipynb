```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
```

```python
f = open('train.csv')
data = csv.reader(f)
header = next(data)

result = []

for row in data:
    if row[5] != 'NA':
        if row[1] == 'open':
            result.append(int(row[5]))

plt.rc('font', family = 'Malgun Gothic')
plt.hist(result, bins = 6, color = 'skyblue', label = '개원중인 병원')
plt.title('병상수에 따른 개원 병원 수')
plt.rcParams["figure.dpi"] = 300
plt.legend()
#print(result)
```

종합병원 : 입원환자 100명 이상 수용 가능(general hospital)
병원 : 입원 환자 30명 이상 100명 미만 수용 가능(hospital)
의원 : 입원 환자 30명 이하 수용 가능(clinic)
한방병원: traditional_hospital
한의원: traditinoal_clinic

```python
dental_clinic = 0
traditional_clinic = 0
traditional_hospital = 0
clinic = 0
hospital = 0
general_hospital = 0
nursing_hospital = 0


f = open('train.csv')
data = csv.reader(f)
header = next(data)

for row in data:
    if row[6] != 'NA':
        if row[1] == 'open':
            
            if row[6] == 'dental_clinic':
                dental_clinic += 1
            elif row[6] == 'traditional_clinic':
                traditional_clinic += 1
            elif row[6] == 'traditional_hospital':
                traditional_hospital += 1
            elif row[6] == 'clinic':
                clinic += 1
            elif row[6] == 'hospital':
                hospital += 1
            elif row[6] == 'general_hospital':
                general_hospital += 1
            else:
                nursing_hospital += 1
    result = [dental_clinic, traditional_clinic, traditional_hospital, clinic, 
              hospital, general_hospital, nursing_hospital] 
print(result)

labels = ['치과', '한의원', '한방병원', '동네의원', '종합병원', '대학병원', '요양병원']

plt.bar(labels, result, color = 'skyblue', label = '개원 중 병원 수')
plt.title('병원 종류 별 개원 중인 병원 수')
plt.legend()
plt.show()
```
