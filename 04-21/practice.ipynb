```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
```

```python
f = open('train.csv')
data = csv.reader(f)
header = next(data)

result = []

for row in data:
    if row[5] != 'NA':
        if row[1] == 'open':
            result.append(int(row[5]))

plt.rc('font', family = 'Malgun Gothic')
plt.hist(result, bins = 6, color = 'skyblue', label = ' ')
plt.title('    ')
plt.rcParams["figure.dpi"] = 300
plt.legend()
#print(result)
```

```python
dental_clinic = 0
traditional_clinic = 0
traditional_hospital = 0
clinic = 0
hospital = 0
general_hospital = 0
nursing_hospital = 0


f = open('train.csv')
data = csv.reader(f)
header = next(data)

for row in data:
    if row[6] != 'NA':
        if row[1] == 'open':
            
            if row[6] == 'dental_clinic':
                dental_clinic += 1
            elif row[6] == 'traditional_clinic':
                traditional_clinic += 1
            elif row[6] == 'traditional_hospital':
                traditional_hospital += 1
            elif row[6] == 'clinic':
                clinic += 1
            elif row[6] == 'hospital':
                hospital += 1
            elif row[6] == 'general_hospital':
                general_hospital += 1
            else:
                nursing_hospital += 1
    result = [dental_clinic, traditional_clinic, traditional_hospital, clinic, 
              hospital, general_hospital, nursing_hospital] 
print(result)

labels = ['', '', '', '', '', '', '']

plt.bar(labels, result, color = 'skyblue', label = '   ')
plt.title('      ')
plt.legend()
plt.show()
```
