## UNIT 13 숫자 데이터를 쉽게 다루게 돕는 numpy 라이브러리

```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
```

```python
# 0 이상 1 이하의 n개의 실수 출력
a = np.random.rand(5)
print(a)
print(type(a))
# 0부터 5까지의 숫자 중 10번을 무작위로 선택
print()
print(np.random.choice(6, 10))
# 0부터 9까지의 숫자 중 6번을 중복없이 무작위로 선택
print()
print(np.random.choice(10, 6, replace=False))
# 각 경우의 수가 발생할 확률을 직접 지정하기
print()
print(np.random.choice(6, 10, p=[0.1, 0.2, 0.3, 0.2, 0.1, 0.1]))
```

```python
# nparray 생성하기
# nparray : N-Dimensional Array, 즉 N차원 배열
a1 = np.array([1,2,3,4]) # 원소가 차례대로 1, 2, 3, 4인 배열
print(a1)
a2 = np.zeros(10)  # 원소가 전부 0이고 크기가 10인 배열
print()
print(a2)
a3 = np.ones(10)   # 원소가 전부 1이고 크기가 10인 배열
print()
print(a3)
a4 = np.eye(3)  # 3행 x 3열의 단위 배열
print()
print(a4)
```

```python
# .arange(a) : a 이하의 숫자 a개를 원소로 오름차순 정렬한 리스트
print(np.arange(3))
# arange(a, b) : a 이상 b 미만의 숫자 (b-a)개를 원소로 오름차순 정렬한 리스트
print(np.arange(3,7))
# arange(a, b, c) : a 이상 b 미만의 숫자 중 a부터 c만큼 커지는 숫자들을 원소로 오름차순 정렬한 리스트
print(np.arange(3,7,2))
# .linspace(a, b, c) : a 이상 b 이하의 c개 구간으로 나눈 실수 데이터
print(np.linspace(1,2,11))
```

```python
# numpy의 다양한 활용
# 1. 연산기호를 통해 모든 배열의 수를 전부 한번에 계산가능
print(np.zeros(10) + 5) 
print(np.zeros(10) - 5)
print(np.zeros(10) * 5)
print(np.ones(10) / 5)
print()
# 2. numpy 자체 함수를 통해 배열의 값 한번에 계산가능
print(np.linspace(1,2,11))
print(np.sqrt(np.linspace(1, 2, 11)))
```

```python
# mask : 특정 조건에 부합하는 데이터를 선별적으로 저장하는 기능
# 특별한 형식은 없지만 일반적으로 조건식의 형태를 띄고 있으면 mask라 생각해도 무관
a = np.arange(-5, 5)
print(a [a < 0]) # a라는 배열 중 0 이하의 원소들만 가져오는 식으로 갱신
# 여기서 'a < 0'가 마스크
# 마스크 두개를 연산기호 +, *를 이용해 논리연산처럼 사용가능
# + : 논리 연산자 or 역할 | * : 논리 연산자 and 역할
mask1 = abs(a) > 3
mask2 = abs(a) % 2 == 0
print(a [mask1 + mask2])   # 둘 중 하나라도 참일 경우
print(a [mask1 * mask2])   # 둘 다 참일 경우
```
