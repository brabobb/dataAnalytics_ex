```python
# numpy 라이브러리 이용한 코드
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
# 파이썬 리스트를 사용한 코드
import matplotlib.pyplot as plt
t = []
p2 = []
p3 = []
for i in range(0, 50, 2) :
    t.append(i / 10)
    p2.append((i / 10) ** 2)
    p3.append((i / 10) ** 3)
plt.plot(t, t, 'r--', t, p2, 'bs', t, p3, 'g^')
plt.show()
```

### numpy array 생성하기
- numpy 라이브러리는 ndarray라는 특별한 데이터타입 배열 사용
- n차원배열이라는 의미를 지는 ndarray 지금부터 array라고 부름

```python
# 리스트와 같이 인덱싱과 슬라이싱 할 수 있다
import numpy as np
a = np.array([1,2,3,4])
print(a[1], a[-1])  # a의 1번 인덱스 값, -1번 인덱스 값 출력
print(a[1:])        # a의 1번 인덱스를 기준으로 슬라이싱 결과 출력
```

```python
# 정수+문자 같이 넣으면 모두 같은 문자 타입으로 변환
import numpy as np
a = np.array([1,2,'3',4])
print(a)
```

```python
# zeros(), ones(), eye() 같은 다양한 함수를 사용해서 배열을 초기화 가능
import numpy as np
a = np.zeros(10)  # 0으로 이루어진 크기가 10인 배열 생성
print(a)

b = np.ones(10)   # 1로 이루어진 크기가 10인 배열 생성
print(b)

c = np.eye(4)  # 3행 x 3열의 단위 배열 생성
print(c)
```

```python
#연속된 숫자로 데이터 생성
import numpy as np
print(np.arange(3))     # arange( ) 함수에 1개 값 입력
print(np.arange(3,7))   # arange( ) 함수에 2개 값 입력
print(np.arange(3,7,2)) # arange( ) 함수에 3개 값 입력
#첫 번째 숫자부터 두 번째 숫자보다 1만큼 작은 정수까지의 범위에서 
#세 번째 숫자만큼의 간격을 둔 숫자가 저장된 배열이 만들어집니다.
```

```python
# numpy 배열에 sin( ) 함수를 적용해 그래프로 나타낸 결과
import matplotlib.pyplot as plt
import numpy as np
a = np.arange(-np.pi, np.pi, np.pi/100)
plt.plot(a, np.sin(a))
plt.show()
```

```python
# sin() 함수와 cos() 함수를 표현
# 6행을 보면 sin() 함수의 평행이동도 쉽게 구현
import matplotlib.pyplot as plt
import numpy as np
a = np.arange(-np.pi, np.pi, np.pi/100)
plt.plot(a, np.sin(a))
plt.plot(a, np.cos(a))
plt.plot(a+np.pi/2, np.sin(a))
plt.show()
```

### 마스크 기능

```python
import numpy as np
a = np.arange(-5, 5)
print(a)

# 상황을 배열 전체에 조건을 적용해서 다음처럼 표현
print(a<0)

mask1 = abs(a) > 3
print(a[mask1])
```

```python
mask1 = abs(a) > 3
mask2 = abs(a) % 2 == 0
print(a[mask1+mask2])   # 둘 중 하나의 조건이라도 참일 경우
print(a[mask1*mask2])   # 두 가지 조건이 모두 참일 경우
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randint(-100, 100, 1000) # 1000개의 랜덤 값 추출
y = np.random.randint(-100, 100, 1000) # 1000개의 랜덤 값 추출
size = np.random.rand(100) * 100
mask1 = abs(x) > 50            # x에 저장된 값 중 절댓값이 50보다 큰 값 걸러 냄
mask2 = abs(y) > 50            # y에 저장된 값 중 절댓값이 50보다 큰 값 걸러 냄
x = x[mask1+mask2]             # mask1과 mask2 중 하나라도 만족하는 값 저장
y = y[mask1+mask2]  # mask1과 mask2 중 하나라도 만족하는 값 저장
plt.scatter(x, y, s=size, c=x, cmap=‘jet’, alpha=0.7)
plt.colorbar()
plt.show()
```
