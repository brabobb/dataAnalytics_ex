## numpy array

```python
#list와 비슷
test = np.array([1, 2, 3, 4])

#','가 없음
print(test)

#nd-array: n-dimensional array(n차원 배열)
print(type(test))

#리스트와 같이 슬라이싱, 인덱싱 가능
print(test[2:])
print(test[-1])

#배열의 데이터는 1가지 타입만 저장 가능
#숫자 + 문자 -> 문자로 저장
dtype1 = np.array([1, '2', '3', '4'])
dtype2 = np.array([1, 2, 3, '4'])
print(dtype1)
print(dtype2)
```

```python
#크기가 10, 0으로 이루어진 배열 생성
test0 = np.zeros(10)
print(test0)

#크기가 10, 1로 이루어진 배열 생성
test1 = np.ones(10)
print(test1)

#3X3 단위 배열 생성
test3 = np.eye(3)
print(test3)
```

```python
#arange는 정수로만 이루어진 배열 생성

#0~9까지의 배열 
print(np.arange(10))

#1~11까지의 배열
print(np.arange(1, 11))

#2~15까지 짝수만으로 이루어진 배열(마지막 숫자는 간격)
print(np.arange(2, 15, 2))
```

```python
#linspace

#1이상 2미만 구간을 0.1 간격으로 쪼개 실수 배열 생성
xx1 = np.arange(1, 2, 0.1)

#1~2 사이를 11개 구간으로 나눈 실수 생성
xx2 = np.linspace(1, 2, 11)

print(xx1)
print(xx2)
```

```python
#벡터와 스칼라의 합?
test0 = np.zeros(10) + 5
print(test0)

#배열 각 원소들을 제곱근함
testl = np.linspace(1, 2, 11)
print(np.sqrt(testl))
```

```python
#sin graph
test_array = np.arange(-np.pi, np.pi, np.pi/100)
plt.plot(test_array, np.sin(test_array))
plt.rcParams["figure.dpi"] = 300
plt.rcParams['axes.unicode_minus'] = False
plt.show()
```

```python
#함수의 평행이동
plt.plot(test_array, np.sin(test_array))
plt.plot(test_array, np.cos(test_array))
plt.plot(test_array + np.pi / 2, np.sin(test_array))
plt.show()
```

```python
#마스크

#-5이하 5 미만인 배열 생성
test = np.arange(-5, 5)
print(test)

#각 원소들의 0보다 작고 큼 여부 확인 = 마스크
print(test < 0)

#마스크를 다시 배열에 적용
#True인 값들만 출력
print(test[test < 0])

#마스크를 변수에 넣고 적용
mask = test > 0
print(test[mask])
```

```python
#abs = 절대값
mask1 = abs(test) > 3
mask2 = abs(test) % 2 == 0
print(test[mask1])
print(test[mask2])

#합집합, or
print(test[mask1 + mask2])

#교집합, and
print(test[mask1 * mask2])
```

```python
#-100~100 사이 1000개의 랜덤값 추출
x = np.random.randint(-100, 100, 1000)
y = np.random.randint(-100, 100, 1000)
#size = np.random.rand(100) * 100

#x, y 절대값이 50보다 큰 값 T
mask1 = abs(x) > 50
mask2 = abs(y) > 50

#합집합
x = x[mask1 + mask2]
y = y[mask1 + mask2]

#size 설정 제외
plt.scatter(x, y, c = x, cmap = 'jet', alpha = 0.7)
plt.colorbar()
plt.show()
```
