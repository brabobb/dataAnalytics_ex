```python
import matplotlib.pyplot as plt
import csv
```

```python
#  본인 생일 이후 서울 최고 기온 밎 최저 기온 꺾은선그래프 표현 코드
f=open('seoul_temperature.csv', 'r', encoding='cp949')
data=csv.reader(f) # csv 모듈의 reader 메소드에 전달해 data라는 csv reader 객체 생성
next(data)
high=[]
low=[]
for row in data:
    if row[-1] != '' : 
        if 1983 <= int(row[0].split('-')[0]):
            if row[0].split('-')[1]=='02' and row[0].split('-')[2]=='14':
                high.append(float(row[-1]))
                low.append(float(row[-2]))

plt.title('highest and lowest temperature in Seoul after birthday')
plt.plot(high, 'r', label='highest temp')
plt.plot(low, 'b', label='lowest temp')
plt.legend()
plt.show()
f.close()
```

```python
#  1월과 8월 이후 서울 최고 기온 히스토그램 표현 코드
f=open('seoul_temperature.csv', 'r', encoding='cp949')
data=csv.reader(f) # csv 모듈의 reader 메소드에 전달해 data라는 csv reader 객체 생성
next(data)
jan=[]
aug=[]
for row in data:
    if row[-1] != '' : 
        if row[0].split('-')[1]=='01':
            jan.append(float(row[-1]))
        if row[0].split('-')[1]=='08':
            aug.append(float(row[-2]))

plt.title('Highest temperature on January and August in Seoul')
plt.hist(jan, bins=100, color='b', label='january temp')
plt.hist(aug, bins=100, color='r', label='august temp')
plt.legend()
plt.show()
f.close()
```

```python
# 8월 1일부터 31일까지 서울 최고 기온 Boxplot 표현 코드
f=open('seoul_temperature.csv', 'r', encoding='cp949')
data=csv.reader(f) # csv 모듈의 reader 메소드에 전달해 data라는 csv reader 객체 생성
next(data)
aug_day=[]
for i in range(31):
    aug_day.append([])
    
for row in data:
    if row[-1] != '' : 
        if row[0].split('-')[1] == '08' : 
            aug_day[int(row[0].split('-')[-1])-1].append(float(row[-1]))

plt.title('Temperature on August in Seoul')
plt.boxplot(aug_day)
plt.show()
f.close()
```
