```python
#1월,8월의 최고기온 데이터 boxplot으로 표현
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = []
jan = []

for row in data :
    month = row[0].split('-')[1]
    if row[-1] != '' :
        if month == '08':
            aug.append(float(row[-1]))
        if month == '01':
            jan.append(float(row[-1]))


plt.boxplot([aug,jan])
plt.show()
```

```python
#1월과 8월의 최고기온 데이터 히스토그램으로 표현
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = []
jan = []

for row in data :
    month = row[0].split('-')[1]
    if row[-1] != '' :
        if month == '08':
            aug.append(float(row[-1]))
        if month == '01':
            jan.append(float(row[-1]))
plt.hist(jan, bins = 100, color='b')
plt.hist(aug, bins = 100, color='r') #측정된 일수는 100일 간격으로, 히스토그램
plt.show()
```

```python
#내 생일(1997.05.21)의 태어난 이후 매년 최고기온 시각화
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)
hot = []
cold = []

for row in data :
    if row[-1] != '' :
        if 1997 <= int(row[0].split('-')[0]) :
            if row[0].split('-')[1] == '05' and row[0].split('-')[2] == '21':
                hot.append(float(row[-1]))
                cold.append(float(row[-2]))
                                  
plt.plot(hot, 'hotpink', label='hot')
plt.plot(cold, 'blue', label = 'cold')
plt.legend()
plt.show()
```
